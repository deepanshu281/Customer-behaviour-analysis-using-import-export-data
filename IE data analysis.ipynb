```python
# Cleaning Data

# Specify the columns to keep
columns_to_keep = ['Date', 'HS Code', 'Product Description', 'Importer Name',
                   'Importer State', 'Exporter Name', 'Standard Qty', 
                   'Standard Unit', 'Port of Destination', 'Shipment Mode', 
                   'Country of Origin', 'Port Of Origin']

# Filter out required products with exact string matching
products_to_filter = ['ORTHOXYLENE', 'PARAXYLENE', 'Vinyl Acetate Monomer', 
                      'Styrene Monomer', 'Butyl acetate']

# Create a filtered DataFrame based on exact matches
filtered_df = df[df['Product Description'].isin(products_to_filter)]

# Keep only the necessary columns
filtered_df_cleaned = filtered_df[columns_to_keep]

# Drop rows with missing values (optional)
filtered_df_cleaned = filtered_df_cleaned.dropna()
filtered_df_cleaned.head()
```
